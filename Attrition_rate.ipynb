```python
location=data["Location"].value_counts()
print(location)

location_new = {
    'Chennai':       7,
    'Noida':         6,
    'Bangalore':     5,
    'Hyderabad':     4,
    'Pune':          3,
    'Madurai':       2,
    'Lucknow':       1,
    'other place':         0,
}

print(location_new)
```

```python
def location(x):
    if str(x) in location_new.keys():
        return location_new[str(x)]
    else:
        return location_new['other place']
    
data_l=data["Location"].apply(location)
data['New Location'] = data_l
data.head()
```

```python
gen = pd.get_dummies(data["Function"])
```

```python
hr = pd.get_dummies(data["Hiring Source"])
```

```python
def Mar(x):
    if str(x) in Marital.keys() and Marital[str(x)] > 100:
        return str(x)
    else:
        return 'other status'
    
data_l =data["Marital Status"].apply(Mar)
data['New Marital'] = data_l
data.head()
```

```python
Mr = pd.get_dummies(data["New Marital"])
```

```python
def Promoted(x):
    if x == 'Promoted':
        return int(1)
    else:
        return int(0)
    
data_l =data["Promoted/Non Promoted"].apply(Promoted)
data['New Promotion'] = data_l
data.head()
```

```python
Emp_new = {
    'B1': 4,
    'B2': 3,
    'B3': 2,
    'other group': 1,
}


def emp(x):
    if str(x) in Emp_new.keys():
        return str(x)
    else:
        return 'other group'
    
data_l =data["Emp. Group"].apply(emp)
data['New EMP'] = data_l

emp = pd.get_dummies(data["New EMP"])
data.head()
```

```python
def Job(x):
    if x == 'Yes':
        return int(1)
    else:
        return int(0)
    
data_l =data["Job Role Match"].apply(Job)
data['New Job Role Match'] = data_l
data.head()
```

```python
def Gen(x):
    if x in gender.keys():
        return str(x)
    else:
        return 'other'
    
data_l =data["Gender "].apply(Gen)
data['New Gender'] = data_l
data.head()
```

```python
gend = pd.get_dummies(data["New Gender"])
tengrp = pd.get_dummies(data["Tenure Grp."])
dataset = pd.concat([data, hr, Mr, emp, tengrp, gen, gend], axis = 1)
```

```python
dataset = pd.read_csv("processed table.csv")
dataset = pd.DataFrame(dataset)
y = dataset["Stay/Left"]
X = dataset.drop("Stay/Left",axis=1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
lr=LogisticRegression(C = 0.1, random_state = 42, solver = 'liblinear')
dt=DecisionTreeClassifier()
rm=RandomForestClassifier()
gnb=GaussianNB()
```

```python
for a,b in zip([lr,dt,rm,gnb],["Logistic Regression","Decision Tree","Random Forest","Naive Bayes"]):
    a.fit(X_train,y_train)
    prediction=a.predict(X_train)
    y_pred=a.predict(X_test)
    score1=accuracy_score(y_train,prediction)
    score=accuracy_score(y_test,y_pred)
    msg1="[%s] training data accuracy is : %f" % (b,score1)
    msg2="[%s] test data accuracy is : %f" % (b,score)
    print(msg1)
    print(msg2)
```

```python
model = lr.fit(X_train,y_train)

import pickle
filename = 'finalized_model.pickle'
pickle.dump(model, open(filename, 'wb'))
```

```python
loaded_model=pickle.load(open(filename,'rb'))
print(loaded_model.score(X_train,y_train))
print(loaded_model.score(X_test,y_test))
```
